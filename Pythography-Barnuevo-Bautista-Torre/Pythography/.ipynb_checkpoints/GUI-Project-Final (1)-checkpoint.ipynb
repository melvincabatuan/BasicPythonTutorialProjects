```python
import tkinter
from tkinter import *
import random

root = Tk()
root.title("GUI Project")
root.geometry("700x600")
root.config(background="#FFDF00")
#########################################################################################################################
#
#
#                                                        BEGINNER
#
#
#########################################################################################################################


def runBeginner():
    global rootBeginner
    
    rootBeginner = Tk()               
    rootBeginner.title("GUI Project") 
    rootBeginner.geometry("700x750")  
    rootBeginner.config(background="#1974D2")
    
    img1 = PhotoImage(file = "img1.png") #Ito para maread niya yung image file
    smallImage = img1.subsample(3,3) #Para paliitin yung size nung image kasi ang laki
    lblImage = Label(rootBeginner, image=smallImage, background="#1974D2") #Para madisplay yung image using Label
    lblImage.pack()

    Lbl1 = Label(rootBeginner,
                text = "For your first lesson, I am going to teach you how to print in Python\n"
                "To do this, we use the command PRINT\n"
                "For example:\n" 
                "The input is: print(\"Hello World!\")\n"
                "The output is: Hello World!\n\n"
                "Now, to test your skill, try and print the word APPLE (all in uppercase). Enter your answer below.\n",font=("Times", 14),
                background = "#FFDF00",foreground="#1974D2"
                )
    Lbl1.pack()
    EntryBox1 = Entry(rootBeginner, width = 50)
    EntryBox1.pack()
    B1 = Button(rootBeginner, text = "Enter", command = exercise1)
    B1.pack()
    B2 = Button(rootBeginner, text = "Next Lesson", command = lesson2, state = DISABLED)
    B2.pack()
    frame1 = Frame(rootBeginner)
    frame1.pack()

def exercise1():#1974D2#FFDF00
    global Lbldef1, Lbldef2, Lbldef3 #etong global di to kailangan wag niyo pansinin
    
    global B2
    
    if EntryBox1.get() == "print" + "(" + "\"APPLE\"" + ")": 
        Lbldef1 = Label(frame1, text = "Nice one!")
        Lbldef1.pack()
        B2.config(state = NORMAL)
    elif EntryBox1.get() == "":
        Lbldef2 = Label(frame1, text = "I couldn't see your answer...")
        Lbldef2.pack()
    else:
        Lbldef3 = Label(frame1, text = "Hmm... Try again, there's a lil bit of error there!")
        Lbldef3.pack()

        
Bimage1 = PhotoImage(file = "beginner1.png") 
BsmallImage1 = Bimage1.subsample(1,1) 
BlblImage1 = Label(rootBeginner, image=BsmallImage1, background="#1974D2") 


def lesson2():
    global Lbldef1, Lbldef2, Lbldef3, Lbl2, Lbl3, Lbl4
    global EntryBox2
    global B3, B4
    global frame2
    
    frame1.destroy()
    lblImage.destroy()
    Lbl1.destroy()
    EntryBox1.destroy()
    B1.destroy()
    B2.destroy()
    
    Lbl2 = Label(rootBeginner,
                text = "For your second lesson, I am going to teach you the different DATA TYPES.\n"
                "There are INTEGERS and FLOAT (numeric data types), TRUE or FALSE (boolean data type), "
                "\nSTRING, LIST, and TUPLE (sequence data types)\n",font=("Times", 12),background = "#FFDF00",foreground="#1974D2"
                )
    Lbl2.pack()
    
    BlblImage1.pack()
    
    Lbl3 = Label(rootBeginner,
                text = "By definition, INTEGERS are the positive and negative numbers.\n"
                "This includes 1, 5, 69, -20, 0, and more.\n"
                "But in Python, there is a special numeric data type called FLOAT.\n"
                "A FLOAT is a number with decimals. This includes 1.34, 8.43, -2.32, and more\n"
                "\nTo determine a data type, we use the syntax TYPE()\n"
                "For instance, if the input is TYPE(5), the output will be INT (or integer)\n"
                "But if the input is TYPE(5.1), the output will be FLOAT",font=("Times", 12),background = "#FFDF00",foreground="#1974D2")
    Lbl3.pack()
    
    Lbl4 = Label(rootBeginner,
                text = "Try it yourself! Enter a value to determine if INTEGER or FLOAT..",font=("Times", 12),background = "#FFDF00",foreground="#1974D2")
    Lbl4.pack()
    
    EntryBox2 = Entry(rootBeginner, width = 50)
    EntryBox2.pack()
    B3 = Button(rootBeginner, text = "Enter", command = exercise2)
    B3.pack()
    B4 = Button(rootBeginner, text = "Next Lesson", command = lesson2_1, state = DISABLED)
    B4.pack()
    frame2 = Frame(rootBeginner)
    frame2.pack()
    


def exercise2():
    global EntryBox2
    global B4
    global frame2
    

    if (float(EntryBox2.get()).is_integer()) == True:
        Lbldef4 = Label(frame2, text = "That is an INTEGER")
        Lbldef4.pack()
        B4.config(state = NORMAL)
    elif (float(EntryBox2.get()).is_integer()) == False:
        Lbldef5 = Label(frame2, text = "That is a FLOAT")
        Lbldef5.pack()
        B4.config(state = NORMAL)
    else:
        Lbldef6 = Label(frame2, text = "I don't recognize that data type, but that's neither INTEGER nor FLOAT")
        Lbldef6.pack()
        
Bimage2 = PhotoImage(file = "beginner2.png") 
BsmallImage2 = Bimage2.subsample(1,1) 
BlblImage2 = Label(rootBeginner, image=BsmallImage2, background="#1974D2") 

def lesson2_1():
    global EntryBox3
    global Lbl5, Lbl6
    global B5, B6
    global frame3
    global Q1
    
    BlblImage1.destroy()
    EntryBox2.destroy()
    Lbl2.destroy()
    Lbl3.destroy()
    Lbl4.destroy()
    frame2.destroy()
    B3.destroy()
    B4.destroy()
    
    Lbl5 = Label(rootBeginner,
                 text = "Another data type is the boolean data type, consisting of TRUE or FALSE values.\n"
                 "TRUE can be equal to 1, while FALSE can be equal to 0.\n"
                 "For example, 5 = 5 should return TRUE while 5 = 5.21 should return FALSE.",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Lbl5.pack()
    
    BlblImage2.pack()
    
    Lbl6 = Label(rootBeginner,
                 text = "Try it yourself! There is an equation given, write True or False",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Lbl6.pack()
    
    Q1 = Label(rootBeginner,
               text = "1092 = 1092",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Q1.pack()
    
    EntryBox3 = Entry(rootBeginner, width = 50)
    EntryBox3.pack()
    B5 = Button(rootBeginner, text = "Enter", command = TrueorFalse)
    B5.pack()
    B6 = Button(rootBeginner, text = "Next Lesson", command = lesson2_2, state = DISABLED)
    B6.pack()
    frame3 = Frame(rootBeginner)
    frame3.pack()
    
def TrueorFalse():
    global EntryBox3
    global B5
    global frame3
    
    if EntryBox3.get() == "True":
        Lbldef7 = Label(frame3, text = "Correct!")
        Lbldef7.pack()
        B6.config(state = NORMAL)
    else:
        Lbldef8 = Label(frame3, text = "Please check your answer")
        Lbldef8.pack()
        
def lesson2_2():
    global EntryBox4
    global Lbl7, Lbl8
    global B7, B8
    global frame4
    
    
    BlblImage2.destroy()
    EntryBox3.destroy()
    Lbl5.destroy()
    Lbl6.destroy()
    B5.destroy()
    B6.destroy()
    frame3.destroy()
    Q1.destroy()
    
    Lbl7 = Label(rootBeginner,
                text = "Lastly are the sequence data types. This includes the STRING, LIST, AND TUPLE.\n"
                "STRINGS are enclosed by quotation marks \"\" \n"
                "LISTS are enclosed by square brackets []\n"
                "TUPLES are enclosed by parentheses ()\n"
                "\nFor example, \"Hello World!\" is a string, [1, 2, 3] is a list, and (1, 2, 3) is a tuple.",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Lbl7.pack()
    Lbl8 = Label(rootBeginner,
                text = "Try it yourself! Input a string, a list, or a tuple.",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Lbl8.pack()
    
    EntryBox4 = Entry(rootBeginner, width = 50)
    EntryBox4.pack()
    B7 = Button(rootBeginner, text = "Enter", command = whatDataType)
    B7.pack()
    B8 = Button(rootBeginner, text = "Next Lesson: Basic Arithmetic", command = Lesson3, state = DISABLED)
    B8.pack()
    frame4 = Frame(rootBeginner)
    frame4.pack()
    
def whatDataType():
    global EntryBox4
    global B8
    global frame4
    
    if EntryBox4.get()[0] == "\"" and EntryBox4.get()[-1] == "\"":
        Lbldef9 = Label(frame4, text = "That is a string")
        Lbldef9.pack()
        B8.config(state = NORMAL)
    elif EntryBox4.get()[0] == "[" and EntryBox4.get()[-1] == "]":
        Lbldef10 = Label(frame4, text = "That is a list")
        Lbldef10.pack()
        B8.config(state = NORMAL)
    elif EntryBox4.get()[0] == "(" and EntryBox4.get()[-1] == ")":
        Lbldef11 = Label(frame4, text = "That is a tuple")
        Lbldef11.pack()
        B8.config(state = NORMAL)
    else:
        Lbldef12 = Label(frame4, text = "I do not recognize that")
        Lbldef12.pack()

Bimage3 = PhotoImage(file = "beginner3.png") 
BsmallImage3 = Bimage3.subsample(1,1) 
BlblImage3 = Label(rootBeginner, image=BsmallImage3, background="#1974D2") 


def Lesson3():
    global Lbl2, Lbl3, Lbl4, Lbl5, Lbl6, Lbl7, Lbl8, Lbl9
    global EntryBox2, EntryBox3, EntryBox4, EntryBox5
    global B3, B4, B5, B6, B7, B8, B9, B10
    global frame2, frame3, frame4, frame5
    global Q1, Q2
    
   
    Lbl7.destroy()
    Lbl8.destroy()
    EntryBox4.destroy()
    B6.destroy()
    B7.destroy()
    B8.destroy()   
    frame4.destroy()
    
    Lbl9 = Label(rootBeginner, text = "In this lesson, you will be able to get numerical values\n"
           "by using the basic arithmetic operations such as \n"
           "ADDITION, SUBTRACTION, MULTIPLICATION, DIVISION, EXPONENTIATION and SQUARE rootBeginnerS.\n\n"
           "The following operators are used:\n"
           "Addition: +\n"
           "Subtraction: -\n"
           "Multiplication: *\n"
           "Division: /\n"
           "Exponentiation: **\n"
           "Square rootBeginners: **0.5\n (Note that raising a number to one-half is equal to its square rootBeginner.)",font=("Times", 12),background = "#FFDF00",foreground="#1974D2")
    Lbl9.pack()
    
    
    BlblImage3.pack()
    
    Q2 = Label(rootBeginner,
              text = "Calculate the hypotenuse of a right triangle\n"
               "with a = 5 and b = 8. Use the Pythagorean Theorem c squared = a squared + b squared (c^2=a^2+b^2)",font=("Times", 12),background = "#FFDF00",foreground="#1974D2")
    Q2.pack()
    
    EntryBox5 = Entry(rootBeginner, width = 50)
    EntryBox5.pack()
    B9 = Button(rootBeginner, text = "Enter", command = basicArithmetic)
    B9.pack()
    B10 = Button(rootBeginner, text = "Next Lesson: Integer Division", command = lesson4, state = DISABLED)
    B10.pack()
    frame5 = Frame(rootBeginner)
    frame5.pack()
    
def basicArithmetic():
    global EntryBox5
    global frame5
    global B10
    
    if EntryBox5.get() == "((5**2)+(8**2))**0.5" or EntryBox5.get() == "((8**2)+(5**2))**0.5": 
        Lbldef13 = Label(frame5, text = "Nice one!")
        Lbldef13.pack()
        B10.config(state = NORMAL)
    elif EntryBox5.get() == "":
        Lbldef14 = Label(frame5, text = "I couldn't see your answer...")
        Lbldef14.pack()
    else:
        Lbldef15 = Label(frame5, text = "Hmm... Try again, there's a lil bit of error there!")
        Lbldef15.pack()
#-------inserrt image for question7--------------

Bimage4 = PhotoImage(file = "beginner4.png") 
BsmallImage4 = Bimage4.subsample(1,1) 
BlblImage4 = Label(rootBeginner, image=BsmallImage4, background="#1974D2") 

def lesson4():
    global Lbl9, Lbl10
    global EntryBox5, EntryBox6
    global frame5, frame6
    global B9, B10, B11, B12
    global Q2, Q3
    
    BlblImage3.destroy()
    Lbl9.destroy()
    Q2.destroy()
    EntryBox5.destroy()
    B9.destroy()
    B10.destroy()
    frame5.destroy()
    
    Lbl10 = Label(rootBeginner, 
                  text = "The division taught to you in the previous topic,\n" 
                 "Basic Arithmetic, the symbol used was a single slash \ \n"
                 "For this lesson, you will learn another type of division, which is integer division.\n"
                 "Integers are whole numbers (without decimals). Similar to this, integer division divides \n"
                 "the numbers but only takes the whole number of the answer and disregards the decimals.\n"
                 "It uses the symbol of double slash “//” \nExample:\n\n",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    
    Lbl10.pack()
    
    BlblImage4.pack()
    
    Q3 = Label(rootBeginner, 
               text = "3 divided by 2 is 1.5. This is the answer when a single slash “/” was used \n"
               "but when integer division was used using the double slash “//” the decimal, .5 was not considered in the output,\n"
               "which is why the output is 1.\n"
               "Recreate the program using integer division",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Q3.pack()
    
    EntryBox6 = Entry(rootBeginner, width = 50)
    EntryBox6.pack()
    B11 = Button(rootBeginner, text = "Enter", command = integerDivision)
    B11.pack()
    B12 = Button(rootBeginner, text = "Next Lesson: Math Module", command = lesson5, state = DISABLED)
    B12.pack()
    frame6 = Frame(rootBeginner)
    frame6.pack()
    
def integerDivision():
    global EntryBox6
    global frame6
    global B12
    
    if EntryBox6.get() == "3//2" or EntryBox6.get() == "3 // 2": 
        Lbldef16 = Label(frame6, text = "1, correct!")
        Lbldef16.pack()
        B12.config(state = NORMAL)
    elif EntryBox6.get() == "":
        Lbldef17 = Label(frame6, text = "I couldn't see your answer...")
        Lbldef17.pack()
    else:
        Lbldef18 = Label(frame6, text = "Hmm... Try again, there's a lil bit of error there!")
        Lbldef18.pack()
#------------------mathmodule------------
Bimage5 = PhotoImage(file = "beginner5.png") 
BsmallImage5 = Bimage5.subsample(1,1) 
BlblImage5 = Label(rootBeginner, image=BsmallImage5, background="#1974D2") 

Bimage6 = PhotoImage(file = "beginner6.png") 
BsmallImage6 = Bimage6.subsample(1,1) 
BlblImage6 = Label(rootBeginner, image=BsmallImage6, background="#1974D2") 



def lesson5():
    global Lbl10, Lbl11
    global EntryBox6
    global frame6, frame7
    global B11, B12,  B14
    global Q3
    
    BlblImage4.destroy()
    Lbl10.destroy()
    Q3.destroy()
    EntryBox6.destroy()
    B11.destroy()
    B12.destroy()
    frame6.destroy()
    
    Lbl11 = Label(rootBeginner, 
                  text = "There are special codes that can be accessed by importing. One of them is the MATH MODULE.\n",font=("Times", 14),background = "#FFDF00",foreground="#1974D2") 
                        
    Lbl11.pack()
    
    BlblImage5.pack()
    BlblImage6.pack()
    
    B14 = Button(rootBeginner, text = "Continue", command = lesson5b, )
    B14.pack()

Bimage7 = PhotoImage(file = "beginner7.png") 
BsmallImage7 = Bimage7.subsample(1,1) 
BlblImage7 = Label(rootBeginner, image=BsmallImage7, background="#1974D2") 

    
def lesson5b():
    global Lbl10, Lbl11
    global EntryBox6, EntryBox7
    global frame6, frame7
    global B11, B12, B13,B14, B14a
    global  Q4
    
    BlblImage5.destroy()
    BlblImage6.destroy()
    B14.destroy()
    
    BlblImage7.pack()
    Q4 = Label(rootBeginner, 
               text = "Try it yourself, using math module, find the square rootBeginner calculation of 28",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Q4.pack()
    
    EntryBox7 = Entry(rootBeginner, width = 50)
    EntryBox7.pack()
    B13 = Button(rootBeginner, text = "Enter", command = mathModule)
    B13.pack()
    B14a = Button(rootBeginner, text = "Next Lesson: Variables", command = lesson6, state = DISABLED)
    B14a.pack()
    frame7 = Frame(rootBeginner)
    frame7.pack()
    
def mathModule():
    global EntryBox7
    global frame7
    global B14
    
    if EntryBox7.get() == "math.sqrt(28)" or EntryBox7.get() == "math.sqrt( 28 )": 
        Lbldef19 = Label(frame7, text = "5.291502622129181, correct!")
        Lbldef19.pack()
        B14a.config(state = NORMAL)
    elif EntryBox7.get() == "":
        Lbldef20 = Label(frame7, text = "I couldn't see your answer...")
        Lbldef20.pack()
    else:
        Lbldef21 = Label(frame7, text = "Hmm... Try again, there's a lil bit of error there!")
        Lbldef21.pack()

Bimage8 = PhotoImage(file = "beginner8.png") 
BsmallImage8 = Bimage8.subsample(1,1) 
BlblImage8= Label(rootBeginner, image=BsmallImage8, background="#1974D2") 


def lesson6():
    global Lbl11, Lbl12
    global EntryBox7, EntryBox8
    global frame7, frame8
    global B13, B14, B15, B16
    global Q4, Q5
    
    BlblImage5.destroy()
    BlblImage6.destroy()
    BlblImage7.destroy()
    Lbl11.destroy()
    Q4.destroy()
    EntryBox7.destroy()
    B13.destroy()
    B14.destroy()
    B14a.destroy()
    frame7.destroy()
    
    Lbl12 = Label(rootBeginner, 
                  text = "You can store a specific value or object with the use of a variable.\n"
                         "You should also bear in mind that all numbers, functions, modules and files are considered as objects. Example:\n",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
                    
    Lbl12.pack()
    
    BlblImage8.pack()
    
    Q5 = Label(rootBeginner, 
               text = "y=?\n\n"
               "y + 222\n"
               "What should be the value of y to get the answer 333?",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Q5.pack()
    
    EntryBox8 = Entry(rootBeginner, width = 50)
    EntryBox8.pack()
    B15= Button(rootBeginner, text = "Enter", command = variables)
    B15.pack()
    B16 = Button(rootBeginner, text = "Next Lesson: Increments", command = lesson7, state = DISABLED)
    B16.pack()
    frame8 = Frame(rootBeginner)
    frame8.pack()
    
def variables():
    global EntryBox8
    global frame8
    global B16
   
    if EntryBox8.get() == "111": 
        Lbldef22 = Label(frame8, text = "Correct!")
        Lbldef22.pack()
        B16.config(state = NORMAL)
    elif EntryBox8.get() == "":
        Lbldef23 = Label(frame8, text = "I couldn't see your answer...")
        Lbldef23.pack()
    else:
        Lbldef24 = Label(rootBeginner, text = "Hmm... Try again, there's a lil bit of error there!")
        Lbldef24.pack()

Bimage9 = PhotoImage(file = "beginner9.png") 
BsmallImage9 = Bimage9.subsample(1,1) 
BlblImage9= Label(rootBeginner, image=BsmallImage9, background="#ffffff") 


Bimage10 = PhotoImage(file = "beginner10.png") 
BsmallImage10 = Bimage10.subsample(1,1) 
BlblImage10= Label(rootBeginner, image=BsmallImage10, background="#1974D2") 

def lesson7():
    global Lbl12, Lbl13,Labell13
    global EntryBox8, EntryBox9
    global frame8, frame9
    global B15, B16, B17, B18
    global Q5,Q6
    
    BlblImage8.destroy()
    Lbl12.destroy()
    Q5.destroy()
    EntryBox8.destroy()
    B15.destroy()
    B16.destroy()
    frame8.destroy()
    
    Lbl13 = Label(rootBeginner, 
                  text = "There is no x++ expression in python, instead use +=, -=, *=.\n"
                  "The expression a += b is equivalent to the expression a = a + b. Example:\n",font=("Times", 14),background = "#FFDF00",foreground="#1974D2"
                       
                        )
    
    Lbl13.pack()
    
    BlblImage9.pack()
    
    Labell13 = Label(rootBeginner, 
                  text ="Starting from the variable x = 0, the value will add 1 to the value of x because of the expression x+=1. ",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Labell13.pack()
    
    Q6 = Label(rootBeginner, 
               text = "What would be the output of the program below?",font=("Times", 14),background = "#FFDF00",foreground="#1974D2")
    Q6.pack()
    
    BlblImage10.pack()
    
    EntryBox9 = Entry(rootBeginner, width = 50)
    EntryBox9.pack()
    B17 = Button(rootBeginner, text = "Enter", command = increment)
    B17.pack()
    B18 = Button(rootBeginner, text = "Next Lesson: If-Else Statements", command = lesson8, state = DISABLED)
    B18.pack()
    frame9 = Frame(rootBeginner)
    frame9.pack()
    
def increment():
    global EntryBox9
    global frame9
    global B18
    
    if EntryBox9.get() == "20": 
        Lbldef25 = Label(frame9, text = "Correct!")
        Lbldef25.pack()
        B18.config(state = NORMAL)
    elif EntryBox9.get() == "":
        Lbldef26 = Label(frame9, text = "I couldn't see your answer...")
        Lbldef26.pack()
    else:
        Lbldef27 = Label(frame9, text = "Hmm... Try again, there's a lil bit of error there!")
        Lbldef27.pack()

Bimage11 = PhotoImage(file = "beginner11.png") 
BsmallImage11 = Bimage11.subsample(1,1) 
BlblImage11= Label(rootBeginner, image=BsmallImage11, background="#1974D2") 
        
def lesson8():
    global Labell13
    global EntryBox9, EntryBox10
    global frame9, frame10
    global B17, B18, B20
    global Q6
    
    
    BlblImage9.destroy()
    BlblImage10.destroy()
    Labell13.destroy()
    Q6.destroy()
    Lbl13.destroy()
    EntryBox9.destroy()
    B17.destroy()
    B18.destroy()
    frame9.destroy()
    
    Lbl14 = Label(rootBeginner,
                  text = "The IF statements will only execute the instruction or code if the condition is true.\n"
"The IF statements should contain the condition statement and end with a colon (if condition:)\n"
"The code is indented on the next line.\n\n"
"The ELSE statement will execute the code if it did not satisfy the first condition (if condition).\n"
"The ELSE statements should contain the else keyword followed by colon (else :).\n"
"Like the if statement, the code is also indented on the next line.\n\n"
"The ELIF statement always follows an if or another elif statement.\n" 
"This statement provides another condition that is checked only if all of the previous conditions were false.\n"
"The ELIF statements should contain the elif keyword, a condition that evaluates to true or false, and end with a colon(:).\n"
"The code is indented on the next line.",font=("Times", 14),background = "#FFDF00",foreground="#1974D2"
                        )
    
    Lbl14.pack()
    
    BlblImage11.pack()
    
    #Q7 = Label(rootBeginner,
               #  text = "e=7\n"
#"f=7\n"
#"\n"
#"if e<f:\n"
#"    print(\"e is less than f\")\n"
#"\n"
#"_______:\n"
#"    print(\"e is equal to f\")\n"
#"\n"
#"else:\n"
#"    print(\"e is greater than f\")\n"
#"\n"
#"output: e is equal to f"
#"\n"
#"Complete the program.(Hint: elif statement)"
#,
#
   #              background = "#ffffff")
#    Q7.pack()
    
    EntryBox10 = Entry(rootBeginner, width = 50)
    EntryBox10.pack()
    B19 = Button(rootBeginner, text = "Enter", command = ifElse)
    B19.pack()
    B20 = Button(rootBeginner, text = "EXIT PROGRAM", command = rootBeginner.destroy, state = DISABLED)
    B20.pack()
    
def ifElse():
    global EntryBox10
    global B20
    
    if EntryBox10.get() == "elif e==f": 
        Lbldef28 = Label(rootBeginner, text = "Correct!")
        Lbldef28.pack()
        B20.config(state = NORMAL)
    elif EntryBox10.get() == "":
        Lbldef29 = Label(rootBeginner, text = "I couldn't see your answer...")
        Lbldef29.pack()
    else:
        Lbldef30 = Label(rootBeginner, text = "Hmm... Try again, there's a lil bit of error there!")
        Lbldef30.pack()      

########################################################################################################################
#
#
#                                                        MAIN
#
#
########################################################################################################################

questions = [
    "Are you taking STEM?",
    "Would you take a computer-related course in college?",
    "Are you interested in programming?",
    "Did you have any programming classes in the past?",
    "Did you find your programming classes difficult in the past?",
    "Do you have any experience in creating programs using Python programming language?",
    "Did you find your past Python programming experience difficult?",
    "Do you search online for codes?",
    "Does watching online tutorials help you become a better programmer?",
]

answersChoice = [
    ["Yes", "No"],
    ["Yes", "No"],
    ["Yes", "No"],
    ["Yes", "No"],
    ["Yes", "No"],
    ["Yes", "No"],
    ["Yes", "No"],
    ["Yes", "No"],
    ["Yes", "No"],
]

answers = [0, 1, 1, 1, 0, 1, 0, 1, 1]
userAnswer = []
indexes = []

def runBEGINNER():

    %run "Beginner.ipynb"
    
def runINTERMEDIATE():
    
    %run "Intermediate.ipynb"
    
def runEXPERT():
    
    %run "Expert.ipynb"
    
def gen():
    global indexes
    while(len(indexes) < 9):
        x = random.randint(0, 8)
        if x in indexes:
            continue 
        else:
            indexes.append(x)

def showresult(score):
    labelQuestion.destroy()
    r1.destroy()
    r2.destroy()
    labelresulttext = Label(root, font=("Agency FB", 20), background="green")
    labelresulttext.pack(pady=(265, 0))
    if score >= 31:
        labelresulttext.configure(text = "You're an ADVANCED programmer")
        advancedButton = Button(root, text="Proceed to ADVANCED tutorial", command=runEXPERT)
        advancedButton.pack()
        
    elif (score >= 16 and score < 30):
        labelresulttext.configure(text = "You're an INTERMEDIATE programmer")
        intermediateButton = Button(root, text="Proceed to INTERMEDIATE tutorial", command=runINTERMEDIATE)
        intermediateButton.pack()
        
    else:
        labelresulttext.configure(text = "You're a BEGINNER programmer")
        beginnerButton = Button(root, text="Proceed to BEGINNER tutorial", command=runBEGINNER)
        beginnerButton.pack()
        
        
def calc():
    global indexes, userAnswer, answers
    x = 0
    score = 0
    for i  in indexes:
        if userAnswer[x] == answers[i]:
            score = score + 5
        x += 1
    print(score)
    showresult(score)  
    
ques = 1
def selected():
    global radioVar, userAnswer
    global labelQuestion, r1, r2
    global ques
    x = radioVar.get()
    userAnswer.append(x)
    radioVar.set(-1)
    if ques < 9:
        labelQuestion.config(text=questions[indexes[ques]])
        r1['text'] = answersChoice[indexes[ques]][0]
        r2['text'] = answersChoice[indexes[ques]][1]
        ques += 1
    else:
        calc()
    
def startQuiz():
    global labelQuestion, r1, r2
    labelQuestion = Label(root, 
                          text=questions[indexes[0]], 
                          font=16, 
                          width=500, 
                          justify="center", 
                          wraplength=400, 
                          background="#1974D2",
                          foreground="#ffffff")
    labelQuestion.pack(pady=(200, 30))
    
    global radioVar
    radioVar = IntVar()
    radioVar.set(-1)
    
    r1 = Radiobutton(root, 
                     text=answersChoice[indexes[0]][0], 
                     value=0, 
                     variable=radioVar, 
                     command=selected, 
                     background="#1974D2")
    r1.pack(pady=5)
    r2 = Radiobutton(root, 
                     text=answersChoice[indexes[0]][1], 
                     value=1, 
                     variable=radioVar, 
                     command=selected, 
                     background="#1974D2")
    r2.pack(pady=5)
    
def startIsPressed():
    labelImage.destroy()
    labelText2.destroy()
    buttonStart.destroy()
    labelRules.destroy()
    gen()
    startQuiz()

def enter(e):
    img3 = PhotoImage(file="nextbutton1.png")
    smallerImage3 = img3.subsample(10, 10)
    buttonStart["image"] = smallerImage3
    buttonStart.image = smallerImage3
def leave(e):
    img2 = PhotoImage(file="nextbutton.png")
    buttonStart["image"] = smallerImage2
    buttonStart.image = smallerImage2
    
img1 = PhotoImage(file="pythonlogo.png")
smallerImage = img1.subsample(4,4)
labelImage = Label(root, 
                   image=smallerImage, 
                   background="#FFDF00")
labelImage.pack(pady=(10, 20))

labelText2 = Label(root,
                   text="Python Questionnaire", 
                   font=("Agency FB", 24), 
                   background="#FFDF00")
labelText2.pack()

img2 = PhotoImage(file="nextbutton.png")
smallerImage2 = img2.subsample(10, 10)
buttonStart = Button(root,
                     image=smallerImage2, 
                     relief=GROOVE, 
                     border=0, 
                     command=startIsPressed,
                     background = "#FFDF00")
buttonStart.pack(pady=(40, 0))
buttonStart.image = smallerImage2
buttonStart.bind("<Enter>", enter)
buttonStart.bind("<Leave>", leave)

labelRules = Label(root, 
                   text="This quiz contains 10 questions which will determine the extent of your programming knowledge\nGood luck!", 
                   width=100, 
                   font=("Times", 10), 
                   background="#1974D2", 
                   foreground="#faca2f")
labelRules.pack(pady=(80, 0))



#----------------------------------------------------------------------------------------------------------#






root.mainloop()
```
