```python
import webbrowser
import os

def openipynb(notebook):
    path = os.getcwd()
    workingdir=path.split("\\")
    getpath="/".join(workingdir)
    final=getpath.split("/",3)
    finalpath=final[len(final)-1]

    url='http://localhost:8888/notebooks/{}/{}.ipynb'.format(finalpath,notebook)
    webbrowser.open(url)
```

```python
def scorer(ipynb):
    try:
        with open('{}.ipynb'.format(ipynb), 'r', encoding='utf-8', errors='ignore') as file:
            text=file.read()

        whole=text.split("\n")
        score=[]
        for x in whole:
            if x.startswith("""      "your"""):
                score.append(x)
            else:
                pass

        split=score[0].split("\\")
        separatedscore=split[0].split(" ")
        intScore=int(separatedscore[len(separatedscore)-1])
        return intScore
    except Exception:
        return 0
```

```python
def scorerprint(ipynb):
    try:
        with open('{}.ipynb'.format(ipynb), 'r', encoding='utf-8', errors='ignore') as file:
            text=file.read()

        whole=text.split("\n")
        score=[]
        for x in whole:
            if x.startswith("""      "Done in printing""") or x.startswith("""      "Personalized Learning""") or x.startswith("""      "Python"""):
                score.append(x)
            else:
                pass

        intScore=len(score)
        return intScore
    except Exception:
        return 0
```
